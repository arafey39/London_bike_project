```python
#specifing the column names that i want to use instead
new_cols_dict = {
    'timestamp' : 'Time',
    'cnt' : 'Count',
    't1' : 'Temp_real_C',
    't2': 'Temp_feels_like_C',
    'hum': 'Humidity_percent',
    'wind_speed': 'Wind_speed_kph',
    'weather_code' : 'Weather',
    'is_holiday': 'Is_holiday',
    'is_weekend': 'Is_weekend',
    'season' : 'Season'
}
#renaming the column names to the ones i want to make the data more clear and understanding
df.rename(new_cols_dict, axis =1, inplace = True)
```

```python
#changing the values in the humidity column to percentage
df.Humidity_percent= df.Humidity_percent / 100
```

```python
# creating a dictionary in which i assigned actual written values to the integers to make more sense of the dat
seasons_dict = {
    '0.0' : 'Spring',
    '1.0' : 'Summer',
    '2.0' : 'Autumn',
    '3.0' : 'winter'
}
# assigining values to the weather column so the first time reader can easily understand the data
weather_dict = {
    '1.0' : 'Clear',
    '2.0' : 'Scattered Clouds',
    '3.0' : 'Broken Clouds',
    '4.0' : 'Cloudy',
    '7.0' : 'Rain',
    '10.0' : 'Rain with thunderstorm',
    '26.0' : 'Snowfall'
}
# changing the season data type from string to integer because we assigned alphabatical values to the season column
df.Season= df.Season.astype('str')
#redirecting the season column to the seasons dictionary column in whuch we assigned the updated values
df.Season = df.Season.map(seasons_dict)

#also changing weather values from string to integer
df.Weather = df.Weather.astype('str')
#mapping the new updated column to the previous weather column
df.Weather = df.Weather.map(weather_dict)
```
